```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#Gathering only the desired columns in teh CSV file 
players_character=purchase_data.loc[:,["Gender","SN","Age"]]
#Printing the restults the assure they are correct
#print(players_character)
#Dropping any dupicated entries
players_character=players_character.drop_duplicates()
#Creating a variable that holds the total count (count is method)
Total_players=players_character.count()[0]
#Placing the collected data into a pd dataframe - simialr to print
pd.DataFrame({"Total Players":[Total_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#sum
Total_Purchase=purchase_data["Price"].sum()
#mean
Mean_Purchase=purchase_data["Price"].mean()
#count
Count_Purchase=purchase_data["Price"].count()
#unique
Unique_item=len(purchase_data["Item ID"].unique())
table_summary = pd.DataFrame({"Total Revenue": [Total_Purchase],"Average Price": [Mean_Purchase], "Number of Purchases": [Count_Purchase], "Number of Unique Items": [Unique_item]})
table_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Total_gender=players_character["Gender"].value_counts()
gender_df=purchase_data
#nuique=Total Count
gender_results = gender_df.groupby('Gender')['SN'].agg(['nunique'])
gender_results['Percentage of Players'] = gender_results['nunique']/gender_results['nunique'].sum()
gender_results['Percentage of Players'] = gender_results['Percentage of Players'].map("{:.2%}".format)
gender_results
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data_df=purchase_data
purchasing_analysis_gender_df = purchase_data_df.groupby(['Gender'])
purchase_count = purchasing_analysis_gender_df['Price'].count()
avg_purchase_price = purchasing_analysis_gender_df['Price'].mean()
total_purchase_value = purchasing_analysis_gender_df['Price'].sum()
avg_total_purchase_perperson = purchase_count/purchasing_analysis_gender_df["SN"].nunique()

purchasing_analysis_gender_final_df = pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":avg_purchase_price,"Total Purchase Value":total_purchase_value,"Average Total Purchase per Person":avg_total_purchase_perperson})
purchasing_analysis_gender_final_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data_df=purchase_data
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
bin_labels = ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data_df["Age_Bins"] = pd.cut(purchase_data_df["Age"],bins,labels=bin_labels,include_lowest=True)
purchasing_analysis_age_group = purchase_data_df.groupby("Age_Bins")
# purchasing_analysis_age_group = purchasing_analysis_gender_df.groupby(['Age_Bins'])
purchasing_analysis_age_group_count = purchasing_analysis_age_group['SN'].nunique()
purch_analy_age_df = pd.DataFrame(purchasing_analysis_age_group_count)
purch_analy_age_df['Percentage of Players'] = purch_analy_age_df['SN']/purch_analy_age_df['SN'].sum()
purch_analy_age_df['Percentage of Players'] = purch_analy_age_df['Percentage of Players'].map("{:.2%}".format)

purch_analy_age_df



```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

most_popular_pre1 = purchase_data_df[['Item ID', 'Item Name', 'Price']]
most_popular_pre1 = most_popular_pre1.groupby(['Item ID','Item Name'])

purchase_count_mp = most_popular_pre1['Price'].count()

item_price_mp = most_popular_pre1['Price'].mean()
total_purchase_value_mp = item_price_mp * purchase_count_mp

most_popular_df = pd.DataFrame({"Purchase Count": purchase_count_mp, "Item Price": item_price_mp, "Total Purchase Value": total_purchase_value_mp})

most_popular_df['Item Price'] = most_popular_df['Item Price'].map("${:,.2f}".format)
most_popular_df['Total Purchase Value'] = most_popular_df['Total Purchase Value'].map("${:,.2f}".format)
most_popular_df.sort_values('Purchase Count', ascending=False).head()
```
